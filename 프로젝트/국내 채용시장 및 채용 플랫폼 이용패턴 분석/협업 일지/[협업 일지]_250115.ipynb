#### Reflection
북마크 주제에 대한 질문 2개를 만들어 분석을 진행하였습니다.

첫 번째 질문 : 북마크 후 실제 지원까지 소요 시간 분석

두 번째 질문 : 북마크 후 공고 페이지 재방문 여부와 전환율 차이 분석

분석도 분석이지만, 분석까지의 과정, 스토리 라인을 탄탄하게 만들기 위해 고민을 많이 했습니다.

첫 번째 질문은 제가 왜 '북마크' 란 기능적 요소를 선택하게 됐는지에 대한 합당한 이유를 만들기 위해 만들었습니다.

실제로 분석을 진행한 결과, 북마크는 사용자가 기업에 지원을 빨리 할 수 있도록 도와주는 요인으로 판단할 수 있었습니다.

두 번째 질문은 첫 번째 질문으로부터 생겨났습니다.

북마크는 사용자의 지원에 도움이 되지만, 왜 사용이 잘 되지 않는지 그 점이 궁금하였습니다.

북마크의 기능이 왜 사용이 잘 안되는지, 기능적인 문제가 있는건지(북마크 사용자가 기능을 온전히 사용하는지)에 대한 분석이 추가적으로 필요하다고 판단했습니다.

북마크 사용 그룹을 대상으로 추가 분석을 진행했고, 재방문 여부와 전환율 사이의 관계를 분석했습니다.

북마크란 기능은 내가 관심있는 공고를 저장해두었다가 다시 방문하기 쉽게 하기 위함인 것으로 생각하여 '재방문' 이라는 사용자의 행동을 분석했습니다.

북마크를 하고 재방문을 하는 행위가 실제 지원 전환율과 관계가 있는지 파악하기 위해 그룹을 나누어 가설검정을 진행했습니다.

분석 결과 북마크 후 공고 페이지 재방문 행동이 지원 전환율에 긍정적인 영향이 있으니, 

사용자가 북마크 기능을 온전히 사용할 수 있도록 개선방안을 제안하면 좋을 것 같다는 생각이 들어

멘토링시간과 다음 팀 회의때 팀원들과 토의하여 구체적인 개선방향을 제시하면 좋을 것 같습니다.
