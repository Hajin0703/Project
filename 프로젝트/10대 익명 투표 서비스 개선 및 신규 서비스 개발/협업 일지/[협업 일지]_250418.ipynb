- DAU/MAU 지표를 확인해 보니 인사이트를 얻을 수 있었고, 다음과 같았다.

| 월                  | DAU/MAU          | 해석                             |
| ------------------ | ---------------- | ------------------------------ |
| 2023-04            | **0.662**        | 런칭 초기, 테스트 또는 프로모션 등으로 활발한 사용  |
| 2023-05            | 0.285            | 어느 정도 안정적인 관심 유지               |
| 2023-06 \~ 2024-01 | **0.05 \~ 0.13** | 점점 **활용도가 낮아짐**, 대부분 유저가 앱을 떠남 |
| 2024-02 \~ 2024-05 | **0.13 → 0.20**  | **약간 회복세**, 하지만 여전히 낮음         |

인사이트
초기(4~5월)는 매우 활발했지만, 시간이 갈수록 유저의 재방문율이 급격히 하락함.

23년 6월부터는 DAU/MAU가 0.1 이하인 달이 많아, 서비스 지속성이 매우 낮은 상황.

최근 2~3달(2024-02 ~ 05)에 소폭 상승한 건 긍정적 신호지만, 여전히 지속적인 이용 서비스로 보기 어려움.
