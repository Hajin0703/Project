#### Reflection
- 전체 가입자 중 질문을 열어보거나 응답을 기록한 유저는 0.7%에 불과함.
- 99% 이상은 가입만 하고, 서비스의 핵심 기능을 사용하지 않는다고 생각함.
- 결론적으로 가입 대비 활동률이 1%도 되지않는다는건 서비스 지속성뿐 아니라 초기 활성화 단계부터 큰 이슈가 있다고 생각하여
이를 고려한 분석을 진행하기로 함.

총 3가지 스텝을 정해놨는데 오늘은 스텝1만 진행함.

Step 1: 사용자 행동 기반 세분화
✅ 분석 목표
"어떤 유형의 유저가 지속성에 기여하지 못하고 있는가?"를 파악해서
→ 서비스에서 우선적으로 타겟팅해야 할 유저 유형을 정의하는 것

✅ 우리가 사용할 주요 파생 변수들
accounts_userquestionrecord 테이블에 기반해서 다음과 같이 그룹 나눌 수 있어:

가입 후 활동 일수 기준

user_id별 활동 일 수: created_at을 날짜 단위로 변환 후 고유 날짜 수 집계

예: 하루만 활동한 유저, 2~3일 활동, 5일 이상 활동 등

총 활동 횟수 기준

user_id별 전체 기록 수

예: 단발성, 소극적, 활발한 유저

활동 간격 기준

첫 활동 ~ 마지막 활동 날짜 사이의 기간 계산

예: 하루에 몰아서 하는 유저 vs 띄엄띄엄 여러 날 활동하는 유저

참여 깊이 기준

열람만 한 유저 (has_read=True, answer_status=null)

응답한 유저 (answer_status=answered)

신고한 유저 (is_reported=True)

| 사용자 유형 | 수 (명) | 비율 (%)    |
| ------ | ----- | --------- |
| 지속 유저  | 4,370 | 90.1% |
| 간헐적 유저 | 328   | 6.8%      |
| 단발성 유저 | 151   | 3.1%      |

표와 같은 결과를 얻었고 다음과 같이 해석할 수 있음.

활동한 유저들 중 대부분이 지속 유저(90%). 이건 "한 번이라도 활동한 유저는 꽤 오랫동안 활동한다"는 뜻.

그러나 전체 가입자 수(약 677,000)에 비하면 활동하는 유저 수(4,849)는 0.7% 정도밖에 안됨.
→ 즉, 대다수 유저는 가입 후 아예 활동을 안함. → 여기서 지속성 문제의 핵심 원인이 보임.
